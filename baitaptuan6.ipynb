```python
 # W6A4 Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key 
#lặp lại, thêm vào list của key đó


s = input().split() 
d={}
for i in s : 
    k,v=i.split(":") 
    if k not in d : 
        d[k] = []
    d[k].append(v) 
    
print(d) 
    
 
```

```python
#W6A5 Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)

lists=list(map(int, input().split())) 
d={"positives": 0 , "negatives": 0 , "zero": 0  
} 

for i in lists : 
    if i > 0 : 
        d["positives"] += 1 
    elif i < 0 : 
        d["negatives"] += 1
    else : 
        d["zero"] += 1
print(d) 



```

```python
#W6A9Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có 
#trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key 






def input_dict():
  return eval(input())

d1 = input_dict()
d2 = input_dict()
d={} 
for i in d1 : 
    d[i] = d1[i] 
    
for k in d2 : 
    if k  in d :
        d[k] += d2[k] 
    else : 
        d[k] = d2[k]
for k in sorted(d) : 
    print(f"{k}:{d[k]}",end=" ")

    
        
        
        

    


    
        
    
        
        
    



 


```

```python
#W6A10!!
nums = list(map(int, input().split()))
k = int(input())

pairs = []
n = len(nums)

for i in range(n):
    for j in range(i + 1, n):
        if nums[i] + nums[j] == k:
            pairs.append((nums[i], nums[j]))

pairs.sort(key=lambda x: x[0])

print(pairs)


```

```python
#6A12!!
nums = list(map(int, input().split()))
nums.sort()
count = {}
for x in nums:
    if x in count:
        count[x] += 1
    else:
        count[x] = 1
max_freq = 0
result = None

for x in count:
    if count[x] > max_freq:
        max_freq = count[x]
        result = x

print(result)
```

```python
#W6A13!!
n = int(input())

d = {}
for _ in range(n):
    key, value = input().split()
    # nếu value là số nguyên
    value = int(value)
    d[key] = value

# đảo key và value
new_dict = {}
for key in d:
    new_dict[d[key]] = key

print(new_dict)

```

```python
#6A15!!
n = int(input())

d = {}
for _ in range(n):
    key, value = input().split()
    value = int(value)
    d[key] = value

k = int(input())

# Lọc dictionary
new_dict = {}

for key in d:
    if d[key] > k:
        new_dict[key] = d[key]

print(new_dict)
```
