```python
#W6A17
n = int(input())

a = []
for _ in range(n):
    row = list(map(int, input().split()))
    a.append(row)

# đường chéo chính
main_diagonal = []
# đường chéo phụ
secondary_diagonal = []

for i in range(n):
    main_diagonal.append(a[i][i])
    secondary_diagonal.append(a[i][n - 1 - i])

print("Duong cheo chinh:", main_diagonal)
print("Duong cheo phu:", secondary_diagonal)
```

```python
#A18
n, m, k = map(int, input().split())

a = []
for _ in range(n):
    a.append(list(map(int, input().split())))

# tổng hàng k
sum_row = 0
for j in range(m):
    sum_row += a[k][j]

# tổng cột k
sum_col = 0
for i in range(n):
    sum_col += a[i][k]

print(sum_row, sum_col)


```

```python
#A19
def index_of_max(nums):
    max_val = nums[0]
    pos = 0

    for i in range(1, len(nums)):
        if nums[i] > max_val:
            max_val = nums[i]
            pos = i

    return pos
a = list(map(int, input().split()))
print(index_of_max(a))
```

```python
#A20
def find_first_index(nums, k):
    for i in range(len(nums)):
        if nums[i] == k:
            return i
    return -1
a = list(map(int, input().split()))
k = int(input())

print(find_first_index(a, k))
```

```python
#A21
s = input().split()

pos = {}

for i, word in enumerate(s):
    if word not in pos:
        pos[word] = []
    pos[word].append(i)

# chuyển list sang tuple
for word in pos:
    pos[word] = tuple(pos[word])

print(pos)
```

```python
#A22
def vi_tri_max(a):
    max_idx = 0
    for i in range(1, len(a)):
        if a[i] > a[max_idx]:
            max_idx = i
    return max_idx
a = [3, 7, 2, 9, 5, 9]
print(vi_tri_max(a))
```

```python
#A23
def xoa_nho_hon_x(a, x):
    i = 0
    while i < len(a):
        if a[i] < x:
            a.pop(i)
        else:
            i += 1
    return a
a = [5, 2, 8, 1, 4, 10]
x = 4

print(xoa_nho_hon_x(a, x))
```

```python
#A24
def day_0_ve_dau(a):
    zeros = []
    others = []
    for x in a:
        if x == 0:
            zeros.append(x)
        else:
            others.append(x)
    return zeros + others

a = list(map(int, input().split()))
ket_qua = day_0_ve_dau(a)
print(*ket_qua)
```

```python
#A25
def so_khong_co(a):
    day_du = set(range(10))
    co_trong_mang = set(a)
    return sorted(day_du - co_trong_mang)


a = list(map(int, input().split()))
ket_qua = so_khong_co(a)

if ket_qua:
    print(*ket_qua)
else:
    print("DAY DU")
```

```python
#A26
def tong_cuc_dai_dia_phuong(a):
    tong = 0
    for i in range(1, len(a) - 1):
        if a[i] > a[i-1] and a[i] > a[i+1]:
            tong += a[i]
    return tong


# ===== Chương trình chính =====
a = list(map(int, input().split()))
print(tong_cuc_dai_dia_phuong(a))
```

```python
#A27
def vi_tri_nho_hon_x(a, x):
    kq = []
    for i in range(len(a)):
        if a[i] < x:
            kq.append(i)
    return kq


a = list(map(int, input().split()))
x = int(input())

vi_tri = vi_tri_nho_hon_x(a, x)

if vi_tri:
    print(*vi_tri)
else:
    print(-1)
```

```python
#A28
def kiem_tra_day(a):
    if len(a) < 2:
        return "VO DANH"

    tang = True
    giam = True

    for i in range(len(a) - 1):
        if a[i] >= a[i + 1]:
            tang = False
        if a[i] <= a[i + 1]:
            giam = False

    if tang:
        return "DAY TANG"
    elif giam:
        return "DAY GIAM"
    else:
        return "VO DANH"



a = list(map(int, input().split()))
print(kiem_tra_day(a))
```
