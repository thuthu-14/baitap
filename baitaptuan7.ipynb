```python
#W7A1 Viết hàm tìm kiếm nhị phân trên một danh sách đã được xắp xếp tăng dần. Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy  
# 1 2 5 6 7 9 

arr = list(map(int, input().split()))
target = int(input())
def  binary_search(arr, left, right, target):
    if left <= right : 
        mid = ( left + right ) // 2 
        if arr[mid] == target : 
            return mid 
        elif arr[mid] >  target : 
            return  binary_search(arr, left, mid - 1, target)
        else : 
            return binary_search(arr, mid + 1, right, target)
    return -1

print(binary_search(arr, 0, len(arr)-1, target))            
            
        
```

```python
#W7A8!!
from bisect import bisect_left, bisect_right

def solve(intervals, queries):
    # lưu query kèm vị trí ban đầu
    qs = [(q, i) for i, q in enumerate(queries)]
    qs.sort()

    sorted_queries = [q for q, _ in qs]
    res = [-1] * len(queries)

    for l, r in intervals:
        length = r - l + 1

        # tìm đoạn query nằm trong [l, r]
        left = bisect_left(sorted_queries, l)
        right = bisect_right(sorted_queries, r)

        for i in range(left, right):
            idx = qs[i][1]
            if res[idx] == -1 or length < res[idx]:
                res[idx] = length

    return res
intervals = [[1,4], [2,4], [3,6], [4,4]]
queries = [2,3,4,5]

print(solve(intervals, queries))




    

                    
                    


                
    
```

```python
#W7A9 Cho một danh sách các số nums, với mỗi phần tử nums[i], hãy đếm xem có bao nhiêu phần tử bên phải nhỏ hơn nums[i].  
#Trả về danh sách các số đếm. Efficient solution: using heap or binary search tree 
from bisect import bisect_left

def count_smaller(nums):
    sorted_list = []
    result = []

    # duyệt từ phải sang trái
    for x in reversed(nums):
        # tìm vị trí chèn (số phần tử < x)
        pos = bisect_left(sorted_list, x)
        result.append(pos)

        # chèn x vào đúng vị trí
        sorted_list.insert(pos, x)

    # đảo lại vì duyệt ngược
    return result[::-1]
nums = [5, 2, 6, 1]
print(count_smaller(nums))


        
        
```

```python
#W7A14
def find_7_index(n,arr):
        index_7=[]
        for i in range (n) :
            if arr[i] == 7 : 
                index_7.append(i)
        return index_7 
     
        
n=int(input())
arr=list(map(int, input().split()))
if n != len(arr) :
    print("Invalid input")
else:
    if len(find_7_index(n,arr)) == 0:
        print("Not found")
    else :
        print(*find_7_index(n,arr))
```
