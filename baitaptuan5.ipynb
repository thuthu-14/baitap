```python
#W5A3 
def songuyento (n) :
    if n<2 :
        return  False 
    for i in range (2,n) :
        if n%i==0 :
            return  False
    return True 

n=int(input())
if songuyento(n):
    print( ' la so nguyen to')
else :
    print( 'khong la so nguyen to')
```

```python
#W5A7!!
def calculator(num1, num2, operat):
    if operat == "+":
        result = num1 + num2
    elif operat == "-":
        result = num1 - num2
    elif operat == "*":
        result = num1 * num2
    elif operat == "/":
        if num2 == 0:
            return "Khong the chia cho 0"
        result = num1 / num2
    else:
        return "Phep toan khong hop le"

    return round(result, 2)


num1 = float(input())
num2 = float(input())
operat = input().strip()

print(calculator(num1, num2, operat))

```

```python
#W5A8
def hammingDistance(x, y):
    return bin(x ^ y).count("1")


x = int(input())
y = int(input())

print(hammingDistance(x, y))

```

```python
#W5A10 
def is_isomorphic(a, b):
    map_ab = {}
    map_ba = {}

    for i in range(len(a)):
        x = a[i]
        y = b[i]

        if x in map_ab:
            if map_ab[x] != y:
                return False
        else:
            map_ab[x] = y

        if y in map_ba:
            if map_ba[y] != x:
                return False
        else:
            map_ba[y] = x

    return True
a = input().strip()
b = input().strip()

print(is_isomorphic(a, b))

```
