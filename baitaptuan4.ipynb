```python
#W4A8
def is_palindrome(n):
    s = str(n)
    return s == s[::-1]

def reverse_number(n):
    return int(str(n)[::-1])

n = int(input())

steps = 0

while not is_palindrome(n):
    n = n + reverse_number(n)
    steps += 1

print(steps, n)


```

```python
#W4A10!!
def collatz_length(n):
    length = 1  
    while n != 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3*n + 1
        length += 1
    return length


n = int(input())

max_len = 0
best_x = 1

for x in range(1, n + 1):
    L = collatz_length(x)

    if L > max_len or (L == max_len and x < best_x):
        max_len = L
        best_x = x

print(best_x, max_len)

```
