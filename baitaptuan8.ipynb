```python
#W7A18
def sx(arr) :
    C=[]
    L=[]
    for i in arr :
        if i % 2 == 0 :
            C.append(i)
        if i % 2 != 0 :
            L.append(i) 
    C.sort()
    L.sort( reverse = True)
    new = C + L 
    return new
arr = [1, 2, 3, 4, 5, 6]
print(*sx(arr))
        
```

```python
#7A19
def merge_arrays(a, b):
    increasing = True
    if len(a) >= 2:
        increasing = a[0] <= a[1]
    elif len(b) >= 2:
        increasing = b[0] <= b[1]

    i = j = 0
    result = []

    while i < len(a) and j < len(b):
        if increasing:
            if a[i] <= b[j]:
                result.append(a[i])
                i += 1
            else:
                result.append(b[j])
                j += 1
        else:  
            if a[i] >= b[j]:
                result.append(a[i])
                i += 1
            else:
                result.append(b[j])
                j += 1

    while i < len(a):
        result.append(a[i])
        i += 1

    while j < len(b):
        result.append(b[j])
        j += 1

    return result
a = [1, 3, 5]
b = [2, 4, 6]
print(merge_arrays(a, b))

```

```python
#W7A25   454878434354    88454 

def max_number_with_k_digits(num: str, k: int) -> str:
    stack = []
    remove = len(num) - k  # số chữ số cần xóa

    for digit in num:
        while stack and remove > 0 and stack[-1] < digit:
            stack.pop()
            remove -= 1
        stack.append(digit)

    # nếu còn dư số cần xóa, xóa từ cuối
    if remove > 0:
        stack = stack[:-remove]

    return ''.join(stack)
print(max_number_with_k_digits("1924", 2))
print(max_number_with_k_digits("4177252841", 4))

    
        
            
```

```python
#W7A27
def is_valid_brackets(s: str) -> bool:
    stack = []
    pairs = {
        ')': '(',
        '}': '{',
        ']': '['
    }

    for ch in s:
        if ch in '({[':
            stack.append(ch)
        elif ch in ')}]':
            if not stack or stack[-1] != pairs[ch]:
                return False
            stack.pop()

    return len(stack) == 0
s = input().strip()
if is_valid_brackets(s):
    print("Hợp lệ")
else:
    print("Không hợp lệ")

```

```python
#WA28
def is_composite(n):
    if n <= 3:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return True
    return False


def longest_composite_sequence(N):
    max_len = 0
    best_seq = []

    current_seq = []

    for x in range(N + 1, N + 2000):
        if is_composite(x):
            current_seq.append(x)
            if len(current_seq) > max_len:
                max_len = len(current_seq)
                best_seq = current_seq[:]
        else:
            current_seq = []

    return best_seq
N = int(input())
result = longest_composite_sequence(N)

print(len(result))
print(result)
```

```python
#WA29
def longest_increasing_subarray(nums):
    if not nums:
        return []

    best_seq = []
    current_seq = [nums[0]]

    for i in range(1, len(nums)):
        if nums[i] > nums[i - 1]:
            current_seq.append(nums[i])
        else:
            if len(current_seq) > len(best_seq):
                best_seq = current_seq[:]
            current_seq = [nums[i]]

    # kiểm tra dãy cuối
    if len(current_seq) > len(best_seq):
        best_seq = current_seq[:]

    return best_seq
nums = list(map(int, input().split()))
result = longest_increasing_subarray(nums)

print(len(result))
print(result)
```

```python
#Wa30
def longest_increasing_subsequence(nums):
    n = len(nums)
    if n == 0:
        return []

    dp = [1] * n
    prev = [-1] * n

    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i] and dp[j] + 1 > dp[i]:
                dp[i] = dp[j] + 1
                prev[i] = j

    # tìm vị trí kết thúc LIS
    max_len = max(dp)
    idx = dp.index(max_len)  # lấy dãy đầu tiên

    # truy vết
    lis = []
    while idx != -1:
        lis.append(nums[idx])
        idx = prev[idx]

    return lis[::-1]
nums = list(map(int, input().split()))
result = longest_increasing_subsequence(nums)

print(len(result))
print(result)

```

```python
#WA31
def smallest_after_one_swap(num_str):
    digits = list(num_str)
    n = len(digits)

    for i in range(n):
        min_digit = digits[i]
        pos = -1

        for j in range(n - 1, i, -1):
            if digits[j] < min_digit:
                # tránh số bắt đầu bằng 0
                if i == 0 and digits[j] == '0':
                    continue
                min_digit = digits[j]
                pos = j

        if pos != -1:
            digits[i], digits[pos] = digits[pos], digits[i]
            break

    return ''.join(digits)
num = input().strip()
print(smallest_after_one_swap(num))
```
