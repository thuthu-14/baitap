```python
#bai20
n=int(input())
A=[int(i) for i in str(n)]
cnt=0
for v in A :
    if v < 2:     
        continue
    flag= True 
    for u in range (2,v):
        if v%u==0 : 
            flag = False 
            break
    if flag :
        cnt+=1 
print(' số chữ số nguyên tố là',cnt)

            
            
```
