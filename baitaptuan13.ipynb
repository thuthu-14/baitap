```python
#W11A4
def angram(path):
    with open(path) as f:
        a = f.readline()
        b = f.readline()
        for i in a:
            if i not in b:
                return False
        return True
angram("abc.txt")
                
```

```python
#11A12
def  maximumProduct(path: str) -> int:
    with open(path) as f:
        n = int(f.readline())
        nums = list(map(int,f.readline().split()))
        sort_nums = sorted(nums, reverse = True)
        if len(sort_nums) >= 3:
            max_nums_1 = sort_nums[0]*sort_nums[1]*sort_nums[2] 
            max_nums_2 = sort_nums[0]*sort_nums[-1]*sort_nums[-2]
            if max_nums_1 >= max_nums_2:
                       return max_nums_1
            else:
                return max_nums_2
print(maximumProduct("11A12.txt"))

                
            
        
```

```python
#11A13
def moveZeroes(path):
    with open(path) as f:
        n = int(f.readline())
        arr = list(map(int, f.readline().split()))
        for i in arr:
            if i == 0 :
                arr.remove(i)
                arr.append(0)
    return arr
print(moveZeroes("11A13.txt"))
            
    
```

```python
#11A14
def averageTime(path):
    A = []
    with open(path)as f:
        lines = f.readlines()
        for i in lines[1:]:
            line = list(map(float, i.split(",")))
            A.append(line[1])
        return sum(A) / len(A)
print(averageTime("11A14.csv"))
            
            
            
        
```

```python
#11A17  6 7 3 4 
def ProductofArrayExceptSelf(path):
    with open(path) as f:
        result = []
        n = f.readline()
        nums = f.readline().strip()
        arr = list(map(int, nums.split()))
        import math
        for i in range(len(arr)):
            c = arr[i]
            arr.remove(arr[i])
            re = math.prod(arr)
            result.append(re)
            arr.insert(i, c)
    return result
print(*ProductofArrayExceptSelf("11A17.txt"))
            
            
            
            
            
            
            
        
```

```python
#11A19!!
import struct

path = input().strip()

with open(path, "rb") as f:
    while True:
        data = f.read(8)      # 1 double = 8 byte
        if not data:
            break
        x = struct.unpack("d", data)[0]
        print(f"{x:.5f}")

```

```python
#11A20!!
def imageSmoother(path: str):
    with open(path) as f:
        M = int(f.readline().strip())
        img = []
        for _ in range(M):
            img.append(list(map(int, f.readline().split())))

    # Ma trận kết quả
    res = [[0]*M for _ in range(M)]

    for i in range(M):
        for j in range(M):
            total = 0
            count = 0

            # duyệt các pixel xung quanh
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    x = i + dx
                    y = j + dy
                    if 0 <= x < M and 0 <= y < M:
                        total += img[x][y]
                        count += 1

            res[i][j] = total // count   # làm tròn xuống

    # In ảnh sau khi làm mịn
    for row in res:
        print(*row)
```
