```python
#W8A1
pi = 3.14
class Cylinder:
    def __init__ (self, r, h):
        self.r = r
        self.h = h
    def dientich(self):
        s = 2*pi*self.h*self.r + 2*pi*self.r**2
        return f"{s:.2f}"
    def thetich(self):
        v = pi*self.r**2*self.h
        return f"{v:.2f}"
p = Cylinder(4,8)
print(p.dientich(), p.thetich())
```

```python
#W8A2
class Date:
    def __init__(self, d, m, y):
        self.d = d
        self.m = m
        self.y = y 
    def is_year(self):
        if self.y % 400 == 0 or ( self.y % 4 == 0 and self.y % 100 !=0 ):
            return True 
        else :
            return False
    def is_valid(self):
        if self.is_year():
            days_in_month = [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            
        else :
            days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if 1 <= self.d <= days_in_month[self.m]:
            return True 
        else :
            return False 
    def next_day(self):
        if self.is_year():
            days_in_month = [0, 31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            
        else :
            days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.d < days_in_month[self.m] :
            self.d +=1 
        else :
            self.d = 1
            if self.m < 12 :
                self.m += 1 
            else :
                self.m = 1
        return f"{self.d}//{self.m}//{self.y}"
d, m, y = map(int, input().split("//")) 
p = Date(d, m, y)
if p.is_valid() :
    print(p.next_day())
else :
    print(" INVALID ")
    
                
            
    
        
        
```

```python
#W8A3
import math
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def check(self):
        if self.x == 0 and self.y == 0 :
            return "goc toa do"
        elif self.y == 0 and self.x != 0 :
            return "nam tren truc hoanh"
        elif self.x == 0 and self.y != 0 :
            return "nam tren truc tung"
    def khoang_cach(self) :
        d = math.sqrt(self.x**2 + self.y**2) 
        return f"{d:.2f}"
x, y = map(int,input().split())
p = Point(x, y)
p.check()
print(p.khoang_cach())
        
        
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.items = []     
    def add_product(self, name, price):
        product = {"name": name, "price": price}
        self.items.append(product)
    def remove_product(self, name):
        for p in self.items :
            if p["name"] == name :
              self.items.remove(p)
              return True
        return False
    def check_items(self):
        if len(self.items) == 0 :
            return True
        else:
            return False
    def sum_price(self):
        t = 0
        for i in self.items:
            t+= i["price"]
        return t
    def show_items(self):
        if self.check_items():
            print("giỏ hàng đang rỗng")
        else:
            for i in self.items:
                print(i)
    def clear_items(self):
        self.items.clear()
c = ShoppingCart()
while True:
    print("\n===== MENU GIỎ HÀNG =====")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm")
    print("4. Kiểm tra giỏ có rỗng không")
    print("5. Tính tổng tiền")
    print("6. Hiển thị sản phẩm")
    print("7. Xóa toàn bộ giỏ hàng")
    print("==========================")

    choice = input("Chọn chức năng (2–7): ")

    if choice == "2":
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá sản phẩm: "))
        c.add_product(name, price)
        print("Đã thêm sản phẩm!")

    elif choice == "3":
        name = input("Nhập tên sản phẩm cần xóa: ")
        if c.remove_product(name ):
            print("Đã xóa sản phẩm.")
        else:
            print("Không tìm thấy sản phẩm!")

    elif choice == "4":
        if c.check_items():
            print("Giỏ hàng rỗng.")
        else:
            print("Giỏ hàng không rỗng.")

    elif choice == "5":
        print("Tổng tiền giỏ hàng là:", c.sum_price())

    elif choice == "6":
        c.show_items()

    elif choice == "7":
        c.clear_cart()
        print("Đã xóa toàn bộ giỏ hàng!")

    else:
        print("Lựa chọn không hợp lệ!")

    # câu hỏi thoát
    exit_choice = input("\nCó muốn thoát khỏi chương trình không? (yes để thoát): ")
    if exit_choice == "yes":
        print("Chương trình kết thúc.")
        break
        
                
        
        
            
        
        
```
